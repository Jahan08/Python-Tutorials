```python
df= DataFrame(np.random.randn(25).reshape(5,5))
df
```

```python
df.loc[1,2] = np.nan
df.loc[2,3] = np.nan
df.loc[4:] = np.nan
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(how="all") #drop the row or column with all NAN value
```

```python
#AXIS
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(thresh=3) #row must contains 3 values to exist
```

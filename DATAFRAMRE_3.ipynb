```python
mask1 = df1["sex"] == "male"
mask2 = df1["age"] > 30

df1[mask1 & mask2] #both condition have to be true
```

# DUPLICATE ROWS

.duplicated()

```python
df.sort_values("FIRST_NAME", inplace = True)
```

```python
df.head()
```

```python
df[df["FIRST_NAME"].duplicated()] #ONLY return the duplicated values
```

```python
df["FIRST_NAME"].duplicated(keep="last") #keep the last as unique and rest of the same name will ebe duplicated
```

```python
m1= ~df["FIRST_NAME"].duplicated(keep=False) # ~ convert all TRUES to FALSES and all FALSES to TRUES
df[m1] # return the unique First name only
```

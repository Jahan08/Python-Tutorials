# LINE 2D plot

```python
np.random.seed(15)
x = np.linspace(1, 40, 10)#random number between 1 and 40
#x = np.arange(1, 40, 10)#array of number between 1 and 40 with 10 difference
print(x)
y1 = x/(np.random.rand(10)*2)
y2 = x/(np.random.rand(10)*3)

line1 = Line2D(x, y1, label="y1")
line2 = Line2D(x, y2, label="y2")

fig, ax = plt.subplots() #take a notebook and turn a page

ax.add_line(line1)
ax.add_line(line2)
ax.plot() # draw on the page
```

```python
np.random.seed(14)
x = list(range(1,6))
y = np.linspace(1, 10, 5)

fig, ax = plt.subplots()

def makeBars(ax, x, y, w):
    for x, y in zip(x, y):
        bar = Rectangle((x, 0), w, y)
        ax.add_patch(bar)
makeBars(ax, x, y, 0.4)
ax.autoscale()

fig.set_size_inches(15, 5)
#fig
```

```python
for r in ax.patches:
    r.set_edgecolor('k')
    r.set_ls('dashed')
    r.set_facecolor('c')
    r.set_lw(2)
    r.set_hatch('||')
fig
```

```python
makeBars(ax, x, y, 0.8)
fig
```

```python
rec_id = range(10)
zorder = reversed(rec_id)

for r, z in zip(rec_id, zorder):
    ax.patches[r].set_zorder(z)
    
ax.set_xlim(1, 6)
ax.set_ylim(1, 11)
fig.savefig("bar1.pdf", dpi=600, transparent=False)
fig
```

# Annotation

```python
from matplotlib.text import Text
from matplotlib.patches import FancyBboxPatch
from matplotlib.text import Annotation

x = list(range(1, 10))
x=y

text1 = Text(x=3, y=4, text='300 K', color="r", size="30", alpha=1.0, family="sans-serif", weight="bold", rotation="45")
text2 = Text(x=5.5, y=6.5, text='340 K', color="g")

annot = Annotation('300 K', xy=(3,4), xytext=(0.5, 0.8), arrowprops=dict(color = "green", arrowstyle="->"), color="white", size="30")

fig, ax = plt.subplots(figsize=(15,5))

t1= ax._add_text(text1)
t1.set_bbox(dict(facecolor='b', boxstyle='darrow, pad=0.1', alpha=0.7, hatch='\\'))

t2= ax._add_text(text2)

an1 = ax._add_text(annot)
an1.set_bbox(dict(facecolor="red", boxstyle='circle, pad=1.5', ls='dashed', lw=2))

arrow=an1.arrow_patch

arrow.set_linewidth(2)
arrow.set_arrowstyle('-[')
arrow.set_connectionstyle('angle, angleA=-90')

def makeFancyBars(ax, x, y, w):
    for x, y in zip(x,y): 
        fbox =FancyBboxPatch((x,0), w, y, boxstyle='square')
        ax.add_patch(fbox)
    
makeFancyBars(ax, x, y, 0.3)
ax.autoscale()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('barchart.csv', index_col="Year")
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
rock  = ax.bar(df.index, "Rock", data=df, label="Rock %")
jazz  = ax.bar(df.index, "Jazz", data=df, label="Jazz %")
ax.legend()
ax.set_xticks(df.index)
fig
```

```python
for p in rock.patches:
    t = ax.annotate(str(p.get_height())+'%', xy=(p.get_x(), p.get_height()+2))
    t.set(color='black', size=10)
    t.set_bbox(dict(facecolor='white', boxstyle='circle', lw=1, edgecolor='red'))

fig    
    
```

# APPLY METHOD

```python
df1=pd.read_csv("jamesbond.csv", index_col="Film")
df1.sort_index(inplace=True)
df1.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + 'Millions!'

df1["Box Office"]= df1["Box Office"].apply(convert_to_string_and_add_millions)
df1.head()
```

```python
df1["Budget"]= df1["Budget"].apply(convert_to_string_and_add_millions)
df1.head()
```

```python
COL = ["Budget", "Box Office", "Bond Actor Salary"]
for col in COL:
    df1[col] = df1[col].apply(convert_to_string_and_add_millions)
    
df1.head()
```

# APPLY METHOD AGAIN

```python
df1=pd.read_csv("jamesbond.csv", index_col="Film")
df1.sort_index(inplace=True)
df1.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4] 
    
    if actor == "Sean Connery":
        return "the best"
    elif actor == "Roger Moore" and budget >40:
        return "enjoyable"
    else:
        return "have no clue"

df1.apply(good_movie, axis = "columns") #want to access row wise
```
